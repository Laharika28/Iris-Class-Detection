**K-Nearest Neighbour Classifier on the IRIS Dataset**

Here I have imported some libraries:

```
import pandas as pd
import numpy as np
```

Here I have imported the dataset i am using for this project

```
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

Here i am seeing the data inside the dataset 'keys'

```
print("Keys of the iris_dataset :\n{}".format(iris_dataset.keys()))
```

Here i am seeing the first 200 characters of the DESCR

```
print(iris_dataset['DESCR'][:200]+"\n......")
```

Here i am seeing the value of the target_names 

```
print("Target Names: {}".format(iris_dataset['target_names']))
```

Here i am seeing the values of the features_names

```
print("Features Names: {}".format(iris_dataset['feature_names']))
```

Here i am seeing the type of the data

```
print("Type of Data:{}".format(type(iris_dataset['data'])))
```

Here i am seeing the Shape of the data

```
print("Shape of Data: {}".format(iris_dataset['data'].shape))
```

Here i am seeing the first 5 rows of the data

```
#Individual rows can be called as samples/instances/datapoints
#And each of the row can be called as attributes/features 
print("First 5 rows of the data : \n{}".format(iris_dataset['data'][:5]))
```

Here i am seeing the type of the target

```
print("Type of Target: {}".format(type(iris_dataset['target'])))
```

Here i am seeing the shape of the target

```
print("Shape of Target: {}".format(iris_dataset['target'].shape))
```

Here i am seeing the value of the target

```
print("Target:{} ".format(iris_dataset['target']))
```

Here I'm spillting the data into 75 and 25 percentages for training and testing purposes respectively

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
                   iris_dataset['data'], iris_dataset['target'], random_state=0)
```

Checking the shape of the X_train,y_train,X_test,y_test

```
print("X_train_shape: {}".format(X_train.shape))
print("y_train_shape: {}".format(y_train.shape))
print("X_test_shape: {}".format(X_test.shape))
print("y_test_shape: {}".format(y_test.shape))
```

Here I'm making a model i.e. making a bot

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

Here I'm training the model by the training data

```
knn.fit(X_train,y_train)
```

Predicting for a new data

```
X_new = np.array([[5,2.9,1,0.2]])

print("X_new shape: {}".format(X_new.shape))
```

```
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Prediction target names: {}".format(
      iris_dataset['target_names'][prediction]))
```

Evaluating the model

```
y_pred = knn.predict(X_test)
print("Test set predictions: \n {}".format(y_pred))
```

Checking out the accuracy of our model

```
print("Test set score: {}".format(np.mean(y_pred == y_test)))
```
